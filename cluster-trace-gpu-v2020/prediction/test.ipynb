```python
from sklearn import cluster

# categ = pd.Categorical(df['inst_num'])
categ = df['inst_num']
# bla = set()
# for i in categ:
#     bla.add(int(i))
# print(bla)
categ = categ.astype('int')
```

```python
kmeans = cluster.KMeans(n_clusters=10, max_iter=50, random_state=2)
kmeans.fit(categ.to_numpy().reshape(-1, 1))
```

```python
labels = kmeans.labels_
clusters = pd.DataFrame(labels, index=df['inst_num'], columns=['clusters'])
clusters.index = clusters.index.astype('int')
cluster_dict = clusters.to_dict()['clusters']
```

```python
cluster_dict
```

```python
from collections import OrderedDict

od = OrderedDict(sorted(cluster_dict.items()))
od
```

```python
od[1] = 6
od[800] = 1
```

```python
current_val = -1
count = 0
test_dict = OrderedDict()
for k, v in od.items():
    if v != current_val:
        current_val = v
        count += 1
    test_dict[k] = count
test_dict
```

```python
adf = pd.DataFrame(test_dict.items(), columns=['num_instances', 'cluster'])
adf.to_csv('datasets/instance_cluster_indices.csv')
```

```python
def merge_one_hot_columns(
    df: pd.DataFrame, 
    columns: List[Union[str, int]], 
    column_name: str, 
    drop_merged: bool = True
    ) -> pd.DataFrame:
    if type(columns[0]) == int:
        columns = df.columns[columns]

    merged_series: pd.Series = df[columns].agg('sum', axis=1)
    merged_series.index.name
    if drop_merged:
        df = df.drop(columns=columns)
    
    df[column_name] = merged_series
    
    return df
    
```
